```python
def generate_elipse(x0, y0, a, b):
    x1 = np.random.uniform(-1, 1) * a
    y1 = np.random.uniform(-1 * np.sqrt((1 - x1 ** 2 / a ** 2) * b ** 2), np.sqrt((1 - x1 ** 2 / a ** 2) * b ** 2))
    x1 += x0
    y1 += y0
    return x1, y1
```

```python
def generate_cone(start_angle):
    dist = np.random.uniform(0, 2 * np.pi)
    offset = np.random.uniform(-1, 1) * dist / np.cos(np.pi / 12) * np.sin(np.pi / 12)
    x1 = dist * np.cos(start_angle) + offset * np.sin(start_angle + np.pi)
    y1 = dist * np.sin(start_angle) + offset * np.cos(start_angle + np.pi)
    return x1, y1
```

```python
def generate_spiral(start_angle):
    dist = np.random.uniform(0, 2 * np.pi)
    offset = np.random.uniform(-1, 1) * dist / np.cos(np.pi / 12) * np.sin(np.pi / 12)
    x1 = dist * np.cos(dist + start_angle) + offset * np.cos(dist + start_angle + np.pi)
    y1 = dist * np.sin(dist + start_angle) + offset * np.sin(dist + start_angle + np.pi)
    return x1, y1
```

```python
def generate_concentric_circles(start_r, fin_r):
    angle = np.random.uniform(-np.pi, np.pi)
    r = np.random.uniform(start_r, fin_r)
    x1 = r * np.cos(angle)
    y1 = r * np.sin(angle)
    return x1, y1
```
